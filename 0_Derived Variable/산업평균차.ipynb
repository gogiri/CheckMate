```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset_kosdaq/상장피처_이상치처리.csv',index_col=0)
df1 = pd.read_csv('../dataset_kosdaq/listed_feature_ver3.0.csv',index_col=0)
```

```python
df
```

```python
df1 = df1[['회계년도','통계청한국표준산업분류코드10차(대분류)','거래소코드']]
df = df.merge(df1,how='left',on=(['회계년도','거래소코드']))
```

```python
columns_of_interest = [
'매출액대비영업현금흐름', '순운전자본비율', '운전자본회전률', '자본분배율', '영업이익대비이자보상배율', '총부채대비영업현금흐름', '매출액순이익률', '자본금순이익률', '부채비율', '순이익증가율', '시장자기자본순이익률', '자기자본순이익률', 'EBITDA(백만원)'
]


grouped_data = df.groupby(['통계청한국표준산업분류코드10차(대분류)', '회계년도'])[columns_of_interest].mean().reset_index()

grouped_data
```

```python
# Renaming the original dataframe and the mean dataframe
df_mean = grouped_data

# Merging the original dataframe with the mean dataframe based on '산업코드' and '회계년도'
df_merged = df.merge(df_mean, how='left', on=['통계청한국표준산업분류코드10차(대분류)', '회계년도'], suffixes=('', '_mean'))

# Calculating the difference between the original values and the mean values for the columns of interest
for column in columns_of_interest:
    df_merged[f'{column}_산업평균차'] = df_merged[column] - df_merged[f'{column}_mean']

df_merged
```

```python
df_merged[['거래소코드','회계년도','통계청한국표준산업분류코드10차(대분류)']]
```

```python
df_final_cleaned = df_merged.drop(columns=[col for col in df_merged.columns if col.endswith('_mean')])

df_final_cleaned
```
